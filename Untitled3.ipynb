```python
from datasets_DDSM import create_train_dataset_DDSM, create_train_loader
import matplotlib.pyplot as plt
import cv2
import numpy as np
import torch
```

```python
DATASET = 'DDSM'
TRAIN_DIR_IMAGES = '../DDSM/ddsm_2015/processed_data_set/'
TRAIN_DIR_LABELS = '../DDSM/ddsm_2015/processed_data_set/'
VALID_DIR_IMAGES = '../DDSM/ddsm_2015/processed_data_set/'
VALID_DIR_LABELS = '../DDSM/ddsm_2015/processed_data_set/'
# Class names.
CLASSES = [
    '__background__',
    'BENIGN',
    'MALIGNANT'

]

# Number of classes (object classes + 1 for background class in Faster RCNN).
NC = 3

# Whether to save the predictions of the validation set while training.
SAVE_VALID_PREDICTION_IMAGES = True
IMAGE_SIZE = (1000, 1500)
BATCH_SIZE= 1
NUM_WORKERS = 1
```

```python
train_dataset = create_train_dataset_DDSM(
    TRAIN_DIR_IMAGES, 
    TRAIN_DIR_LABELS,
    IMAGE_SIZE, 
    CLASSES,

)
train_loader = create_train_loader(
    train_dataset, BATCH_SIZE, NUM_WORKERS,
)
for images, targets in train_loader:
    for i, img in enumerate(images):
        if len(targets[i]['boxes']>0):
            xmin, ymin, xmax, ymax = targets[i]['boxes'][0]
            img = img.permute(1,2,0).numpy().copy()
            img =cv2.rectangle(img = (img*255).astype(np.uint8), pt1= (int(xmin), int(ymin)), pt2= (int(xmax), int(ymax)),color = (255,0,0),thickness= 4)
            plt.imshow(img)
            plt.show()
```
